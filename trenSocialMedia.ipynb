```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
# Set path ke ChromeDriver
service = Service('chromedriver-win64\chromedriver.exe')

# Inisialisasi browser
driver = webdriver.Chrome(service=service)

# 1. Buka halaman Instagram dan login
driver.get('https://www.instagram.com/accounts/login/')
time.sleep(3)  # Tunggu beberapa detik hingga halaman login termuat
```

```python
# Buka halaman postingan
driver.get('https://www.instagram.com/p/C-Aof4tzmhW/')
time.sleep(5)  # Tunggu beberapa detik hingga halaman termuat
```

```python
# Fungsi untuk mensimulasikan scrolling ke bawah halaman
def scroll_to_load_comments(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll ke bawah halaman
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Tunggu beberapa detik untuk memuat komentar tambahan
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Scroll untuk memuat lebih banyak komentar
scroll_to_load_comments(driver)
print("Scrolling selesai")

# Ambil komentar menggunakan XPath yang disalin
comments_elements = driver.find_elements(By.XPATH, '//*[@id="mount_0_0_Sv"]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/section/main/div/div[1]/div/div[2]/div/div[2]')

comments = []
for comment in comments_elements:
    comments.append(comment.text)
    print(comment.text)  # Cetak komentar untuk verifikasi

# Simpan komentar ke dalam file
with open('comments.txt', 'w', encoding='utf-8') as f:
    for comment in comments:
        f.write(comment + '\n')

print("Comments saved to comments.txt")

# Tutup browser
driver.quit()
```

```python
# Baca komentar dari file
with open('comments.txt', 'r', encoding='utf-8') as f:
    comments = f.readlines()

# Bersihkan komentar (misalnya, hapus whitespace)
comments = [comment.strip() for comment in comments if comment.strip()]

# Buat DataFrame
komentar = pd.DataFrame(comments, columns=['Comment'])
```
